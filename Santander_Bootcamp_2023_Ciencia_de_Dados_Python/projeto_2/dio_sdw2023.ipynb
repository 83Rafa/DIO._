## Transform
- Padronizar os dados para obter uma melhor compreensão das informações

```python
# exibe os primeiros 3 valores da coluna 'account'
df['account'][0:3]
```

- Os dados de account podem ser divididos em várias colunas: "*id*", "*number*", "*agency*", "*balance*" e "*Limit"*

```python
# converte os valores da coluna 'account' em dicionários
df['account'] = df['account'].apply(ast.literal_eval)

# criação de novas colunas usando os valores correspondentes dos dicionários na coluna 'account'
df['id'] = df['account'].apply(lambda x: x['id'])
df['number'] = df['account'].apply(lambda x: x['number'])
df['agency'] = df['account'].apply(lambda x: x['agency'])
df['balance'] = df['account'].apply(lambda x: x['balance'])
df['limit'] = df['account'].apply(lambda x: x['limit'])

# remove a coluna 'account'
df = df.drop('account', axis=1)

display(df)
```

- Os dados de cartão estavam sem qualquer padão, deixando um pouco confuso o entendimento

```python
# imprimi o primeiro item da coluna 'card'
df['card'][0]
```

- Essas informações também podem ser separadas em colunas distintas: "*card_id*", "*card_number*" e "*card_limit*"

```python
# Converte os valores da coluna 'card' em dicionários
df['card'] = df['card'].apply(ast.literal_eval)

# Cria novas colunas usando os valores correspondentes dos dicionários na coluna 'card'
df['card_id'] = df['card'].apply(lambda x: x['id'])
df['card_number'] = df['card'].apply(lambda x: x['number'])
df['card_limit'] = df['card'].apply(lambda x: x['limit'])

# Remove a coluna 'card'
df = df.drop('card', axis=1)

display(df)
```

```python
# Classifica o DataFrame pelo valor da coluna 'id'
df = df.sort_values('id')

# Exibe o resultado final
display(df)
```

- Criação e padronização de dados das colunas "*number*", e "*digit*"

```python
# divide a coluna 'number' em duas novas colunas 'number' e 'digit', usando o caractere '-' como separador
df[['number', 'digit']] = df['number'].str.split('-', expand=True)

# remove todos os caracteres não numéricos da coluna 'number'0
df['number'] = df['number'].str.replace(r'\D+', '', regex=True)

# substitui valores vazios na coluna 'number' por '1248'
df.loc[df['number'] == '', 'number'] = '1248'

# substitui valores nulos na coluna 'digit' por 0
df['digit'] = df['digit'].fillna(0)
```

- Remoção de dados estranhos nas colunas que seguem

```python
# remove todos os caracteres não numéricos da coluna 'agency'
df['agency'] = df['agency'].str.replace(r'\D+', '', regex=True)

# substitui valores vazios na coluna 'agency' por '1010'
df.loc[df['agency'] == '', 'agency'] = '1010'
```

```python
# padroniza a coluna 'card_number' com '**** **** **** ' + os quatro últimos dígitos
df['card_number'] = '**** **** **** ' + df['card_number'].str.extract(r'(\d{4})\D*$', expand=False)

# substitui valores nulos na coluna 'card_number' por 'Não Possui Cartão'
df.loc[df['card_number'].isnull(), 'card_number'] = 'Não Possui Cartão'
```

```python
df.columns
```

```python
# define a ordem das colunas
cols = ['id', 'name', 'number', 'digit', 'agency', 'balance', 'limit', 'card_id',
       'card_number', 'card_limit', 'news']

# reordena as colunas na ordem especificada pela lista `cols`
df = df[cols]
```

```python
"""Define o valor da coluna 'card_limit' como 0 se o valor correspondente na coluna 'card_number'
for igual a 'Não Possui Cartão', caso contrário, mantém o valor original da coluna 'card_limit'"""

df['card_limit'] = df.apply(lambda x: 0 if x['card_number'] == 'Não Possui Cartão' else x['card_limit'], axis=1)
```

```python
# renomeia as colunas 'number' e 'digit' para 'account_number' e 'acc_digit', respectivamente
df = df.rename({'number':'account_number', 'digit':'acc_digit'}, axis='columns')
df
```

### Gerar os ids filtrados e acessar o JSON com dados dos usuários

- Filtra clientes com balaço de conta igual ou superior a 100.000 para enviar mensagens personalizadas sobre investimentos

```python
def extract_ids_and_names(df):
    # filtra linhas com balance >= 100000
    df_filtered = df[df['balance'] >= 100000]

    # extrai ids, nomes, saldos, e mensagens dos clientes
    ids = df_filtered['id'].tolist()
    names = df_filtered['name'].tolist()
    balance = df_filtered['balance']
    news = df_filtered['news'].tolist()

    # cria o dicionário de saída
    user = {'id': ids, 'name': names, 'balance': balance, 'news': news}

    return user


users = extract_ids_and_names(df)
users
```

```python
# Repositório da API: https://github.com/digitalinnovationone/santander-dev-week-2023-api
sdw2023_api_url = "https://sdw-2023-prd.up.railway.app"
```

```python
# extrai os ids de interesse
user_ids = df_filtered['id'].tolist()
user_ids
```

```python
def get_user(id):
  response = requests.get(f"{sdw2023_api_url}/users/{id}")
  return response.json() if response.status_code == 200 else None


users = [user for id in user_ids if (user := get_user(id)) is not None]
print(json.dumps(users, indent=2))
```

### Acessar a API do ChatGPT
- Acessar a API e enviar as mensagens personalizadas

```python
import openai
from chave import SECRET
```

```python
openai_api_key = SECRET
```

```python
openai.api_key = openai_api_key

def generate_ai_news(user):
    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "system",
                "content": "Você é um especialista em investimentos."
            },
            {
                "role": "user",
                "content": f"Crie uma mensagem para {user}, chame-o pelo nome, sobre o futuro e investimentos do Santander. (até 150 caracteres)"
            }
        ]
    )
    return completion.choices[0].message.content.strip('\"')


for user in users:
  news = generate_ai_news(user)
  print(news)
  user['news'].append({
      "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
      "description": news
  })
```

## Load

```python
# atualização dos usuários na API da SDW2023 com os dados enriquecidos
def update_user(user):
  response = requests.put(f"{sdw2023_api_url}/users/{user['id']}", json=user)
  return True if response.status_code == 200 else False

for user in users:
  success = update_user(user)
  print(f"User {user['name']} updated: {success}")
```
